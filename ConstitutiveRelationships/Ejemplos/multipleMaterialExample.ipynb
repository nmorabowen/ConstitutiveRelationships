```python
# importamos dependencias

import ConstitutiveRelationships as cr
from baseUnits import kgf, cm, ksi, mm
from plotApeConfig import set_default_plot_params, blueAPE
set_default_plot_params()
import matplotlib.pyplot as plt
```

```python
A572=cr.uniaxialBilinealSteel('A572', fy=50*ksi, fsu=1.1*50*ksi, color='k')
A706Gr60=cr.uniaxialBilinealSteel('A706Gr60', fy=60*ksi, fsu=1.25*60*ksi, color='grey')
fc240=cr.uniaxialUnconfinedConcrete('fc240', 240*kgf/cm**2, color='red')
fc280=cr.uniaxialUnconfinedConcrete('fc280', 280*kgf/cm**2, color='green')

fc240c = cr.uniaxialConfinedConcrete('fc240c',
                                          fco=240 * kgf / cm**2,
                                          eco=0.003,
                                          b=30 * cm,
                                          h=40 * cm,
                                          rec=3 * cm,
                                          num_var_b=3,
                                          num_var_h=4,
                                          phi_longitudinal=16 * mm,
                                          num_est_perpendicular_b=3,
                                          num_est_perpendicular_h=4,
                                          phi_estribo=10 * mm,
                                          esu_estribo=0.12,
                                          s=10 * cm,
                                          fye=4200 * kgf / cm**2,
                                          plot=True,
                                          color='k')

fig, ax = plt.subplots(figsize=(10,5))
A572.plot(ax=ax)
A706Gr60.plot(ax=ax)
fc240.plot(ax=ax)
fc280.plot(ax=ax)
fc240c.plot(ax=ax)

fig, ax = plt.subplots(figsize=(10,5))
fc240.plot(ax=ax)
fc240c.plot(ax=ax)
```
